```python
'''
탐색적 데이터 분석 (EDA)


정형화된 데이터의 요소
정형화된 데이터는 행과 열로 구성된 테이블 형태의 데이터를 의미합니다. 각 행은 개별 관측치를 나타내고, 각 열은 변수(특성)를 나타냅니다.
'''

import pandas as pd
import numpy as np

# 예시 데이터
data = {
    'PatientID': [1, 2, 3, 4, 5],
    'Age': [25, 45, 35, 50, 23],
    'Gender': ['남', '여', '남', '여', '남'],
    'BloodPressure': [120, 130, 125, 135, 110],
    'Cholesterol': [200, 220, 210, 230, 190],
    'Diabetes': [0, 1, 0, 1, 0]
}
df = pd.DataFrame(data) # 실제로는 여기에 그냥 파일 경로입력
print(df)
```

```python
'''
테이블 데이터
테이블 데이터는 데이터프레임 형태로 저장됩니다. 데이터프레임은 판다스(Pandas) 라이브러리를 사용하여 생성하고 조작할 수 있습니다.
'''

# 데이터프레임의 기본 정보 확인
print(df.info())

# 데이터프레임의 통계 요약
print(df.describe())
```

```python
'''
1.3 위치 추정
위치 추정은 데이터의 중심 경향을 나타내는 통계량을 계산하는 것을 의미합니다. 대표적인 위치 추정량으로는 평균, 중앙값, 최빈값이 있습니다.
'''

# 평균
print("평균 나이:", df['Age'].mean())
print("평균 혈압:", df['BloodPressure'].mean())
print("평균 콜레스테롤:", df['Cholesterol'].mean())

# 중앙값
print("중앙값 나이:", df['Age'].median())
print("중앙값 혈압:", df['BloodPressure'].median())
print("중앙값 콜레스테롤:", df['Cholesterol'].median())

# 최빈값
print("최빈값 나이:", df['Age'].mode()[0])
print("최빈값 혈압:", df['BloodPressure'].mode()[0])
print("최빈값 콜레스테롤:", df['Cholesterol'].mode()[0])
```

```python
'''
변이 추정
변이 추정은 데이터의 분산 정도를 나타내는 통계량을 계산하는 것을 의미합니다. 대표적인 변이 추정량으로는 분산과 표준편차가 있습니다.
'''

# 분산
print("나이 분산:", df['Age'].var())
print("혈압 분산:", df['BloodPressure'].var())
print("콜레스테롤 분산:", df['Cholesterol'].var())

# 표준편차
print("나이 표준편차:", df['Age'].std())
print("혈압 표준편차:", df['BloodPressure'].std())
print("콜레스테롤 표준편차:", df['Cholesterol'].std())
```

```python
'''
1.5 데이터 분포 탐색하기
데이터의 분포를 시각화하여 탐색할 수 있습니다. 히스토그램, 박스플롯 등을 사용하여 데이터의 분포를 확인합니다.
'''

import matplotlib.pyplot as plt

# 히스토그램
df['Age'].hist()
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

df['BloodPressure'].hist()
plt.title('Blood Pressure Distribution')
plt.xlabel('Blood Pressure')
plt.ylabel('Frequency')
plt.show()

df['Cholesterol'].hist()
plt.title('Cholesterol Distribution')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.show()

# 박스플롯
df.boxplot(column=['Age', 'BloodPressure', 'Cholesterol'])
plt.title('Age, Blood Pressure, Cholesterol Distribution')
plt.show()
```

```python
'''
이진 데이터와 범주 데이터 탐색하기
이진 데이터와 범주 데이터는 각각 두 가지 값과 여러 범주로 구성된 데이터를 의미합니다. 이 데이터의 빈도수를 확인할 수 있습니다.
'''

# 성별 빈도수
print(df['Gender'].value_counts())

# 당뇨 여부 빈도수
print(df['Diabetes'].value_counts())
```

```python
'''
1.7 상관관계
상관관계는 두 변수 간의 관계를 나타내는 통계량입니다. 상관계수는 -1에서 1 사이의 값을 가지며, 1에 가까울수록 강한 양의 상관관계를, -1에 가까울수록 강한 음의 상관관계를 의미합니다.
'''
# 범주형 변수와 'PatientID' 제외
df_numeric = df.drop(columns=['PatientID', 'Gender'])

# 상관계수 계산
correlation_matrix = df_numeric.corr()
print(correlation_matrix)
```

```python
'''
두 개 이상의 변수 탐색하기
두 개 이상의 변수를 동시에 탐색하여 변수 간의 관계를 시각화할 수 있습니다. 산점도 등을 사용하여 변수 간의 관계를 확인합니다.
'''

# 산점도
df.plot.scatter(x='BloodPressure', y='Cholesterol')
plt.title('Relationship between Blood Pressure and Cholesterol')
plt.xlabel('Blood Pressure')
plt.ylabel('Cholesterol')
plt.show()
```

```python
'''
데이터와 표본분포


임의표본추출과 표본편향
임의표본추출은 모집단에서 무작위로 표본을 선택하는 방법입니다.
표본편향은 표본이 모집단을 대표하지 못하는 경우를 의미합니다.
'''

import numpy as np

# 모집단 생성
population = np.random.randint(1, 101, 1000)

# 임의표본추출
sample = np.random.choice(population, 100, replace=False)
print("표본 평균:", np.mean(sample))
print("모집단 평균:", np.mean(population))
```

```python
'''
선택편향
선택편향은 표본이 특정 조건에 의해 선택되어 모집단을 대표하지 못하는 경우를 의미합니다.
'''

# 선택편향 예제
biased_sample = population[population > 50]
print("편향된 표본 평균:", np.mean(biased_sample))
```

```python
'''
통계학에서의 표본분포
표본분포는 동일한 모집단에서 여러 표본을 추출하여 계산한 통계량의 분포를 의미합니다.
'''

# 여러 표본의 평균 계산
sample_means = [np.mean(np.random.choice(population, 100, replace=False)) for _ in range(1000)]
print("표본 평균의 평균:", np.mean(sample_means))
print("표본 평균의 표준편차:", np.std(sample_means))
```

```python
'''
부트스트랩
부트스트랩은 표본에서 반복적으로 데이터를 추출하여 통계량의 분포를 추정하는 방법입니다
'''

# 부트스트랩 예제
bootstrap_samples = [np.mean(np.random.choice(sample, 100, replace=True)) for _ in range(1000)]
print("부트스트랩 표본 평균의 평균:", np.mean(bootstrap_samples))
print("부트스트랩 표본 평균의 표준편차:", np.std(bootstrap_samples))
```

```python
'''
신뢰구간
신뢰구간은 모집단의 모수를 포함할 것으로 예상되는 범위를 의미합니다.
'''

import scipy.stats as stats

# 신뢰구간 계산
confidence_interval = stats.t.interval(0.95, len(sample)-1, loc=np.mean(sample), scale=stats.sem(sample))
print("95% 신뢰구간:", confidence_interval)
```

```python
'''
정규분포
정규분포는 데이터가 평균을 중심으로 대칭적으로 분포하는 형태를 의미합니다.
'''

# 정규분포 예제
mu, sigma = 0, 0.1
s = np.random.normal(mu, sigma, 1000)

# 히스토그램
plt.hist(s, bins=30, density=True)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
'''
긴 꼬리 분포
긴 꼬리 분포는 데이터의 일부가 평균에서 멀리 떨어져 있는 분포를 의미합니다.
'''

# 긴 꼬리 분포 예제
s = np.random.lognormal(0, 1, 1000)

# 히스토그램
plt.hist(s, bins=30, density=True)
plt.title('Long-Tailed Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
'''
스튜던트의 t 분포
스튜던트의 t 분포는 표본 크기가 작은 경우에 사용되는 분포입니다.
'''

# t 분포 예제
s = np.random.standard_t(10, size=1000)

# 히스토그램
plt.hist(s, bins=30, density=True)
plt.title("Student's t Distribution")
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
'''
이항분포
이항분포는 성공과 실패 두 가지 결과를 가지는 실험의 성공 횟수를 나타내는 분포입니다.

동전 던지기: 10번 던졌을 때 앞면이 나오는 횟수.
시험 합격: 20명의 학생 중 15명이 시험에 합격할 확률.
제품 불량: 100개의 제품 중 불량품의 수.
'''

import numpy as np
import matplotlib.pyplot as plt

# 이항분포 예제
n, p = 10, 0.5  # n: 시행 횟수, p: 성공 확률
s = np.random.binomial(n, p, 1000)  # 1000개의 표본 생성

# 히스토그램
plt.hist(s, bins=30, density=True)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```

```python
'''
카이제곱분포
카이제곱분포는 독립적인 정규분포를 따르는 변수들의 제곱합의 분포입니다.

적합도 검정: 관찰된 데이터가 기대되는 분포와 일치하는지 검정.
독립성 검정: 두 범주형 변수 간의 독립성을 검정.
분산 분석: 여러 그룹 간의 분산을 비교.
'''

# 카이제곱분포 예제
df = 2  # 자유도
s = np.random.chisquare(df, 1000)  # 1000개의 표본 생성

# 히스토그램
plt.hist(s, bins=30, density=True)
plt.title('Chi-Square Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
'''
F 분포
F 분포는 두 카이제곱분포를 따르는 변수들의 비율의 분포입니다.

분산 분석(ANOVA): 여러 그룹 간의 평균 차이를 검정.
회귀 분석: 회귀 모델의 적합성을 검정.
비율 검정: 두 모집단의 분산 비율을 비교
'''

# F 분포 예제
dfn, dfd = 2, 2  # dfn: 분자 자유도, dfd: 분모 자유도
s = np.random.f(dfn, dfd, 1000)  # 1000개의 표본 생성

# 히스토그램
plt.hist(s, bins=30, density=True)
plt.title('F Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
'''
푸아송 분포
푸아송 분포는 단위 시간 또는 공간에서 발생하는 사건의 수를 나타내는 분포입니다.

콜센터: 특정 시간 동안 콜센터에 걸려오는 전화의 수.
교통사고: 특정 도로 구간에서 일정 시간 동안 발생하는 교통사고의 수.
웹사이트 방문자: 특정 시간 동안 웹사이트에 접속하는 방문자의 수.
'''

import numpy as np
import matplotlib.pyplot as plt

# 푸아송 분포 예제
lam = 3  # lam: 평균 발생률 (λ)
s = np.random.poisson(lam, 1000)  # 1000개의 표본 생성

# 히스토그램
plt.hist(s, bins=30, density=True)
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.show()
```

```python
'''
지수 분포
지수 분포는 사건 간의 시간 간격을 모델링하는 데 사용됩니다. 예를 들어, 고객이 매장에 도착하는 시간 간격을 모델링할 때 사용됩니다

고객 도착 시간: 매장에 도착하는 고객 간의 시간 간격.
기계 고장 시간: 기계가 고장 나기까지의 시간 간격.
전화 통화 시간: 콜센터에서 전화 통화가 끝난 후 다음 전화가 걸려오기까지의 시간 간격.
'''

# 지수 분포 예제
scale = 1 / 3  # scale: 스케일 매개변수 (λ의 역수), λ는 사건이 발생하는 평균 비율을 나타내며, scale은 그 역수로 사건 간의 평균 시간 간격
s = np.random.exponential(scale, 1000)  # 1000개의 표본 생성

# 히스토그램
plt.hist(s, bins=30, density=True)
plt.title('Exponential Distribution')
plt.xlabel('Time Interval')
plt.ylabel('Frequency')
plt.show()
```

```python
'''
감마 분포
감마 분포는 여러 지수 분포의 합으로 나타낼 수 있으며, 대기 시간의 합을 모델링하는 데 사용됩니다.

프로젝트 완료 시간: 여러 작업 단계를 거쳐 프로젝트가 완료되기까지의 총 시간.
보험 청구: 보험 청구가 처리되기까지의 시간.
수명 분석: 제품이나 부품의 수명.
'''

import numpy as np
import matplotlib.pyplot as plt

# 감마 분포 예제

# shape: 모양 매개변수 (k) 값 클수록 오른쪽으로
# scale: 스케일 매개변수 (θ) 값이 클수록 분포가 넓어짐
shape, scale = 2, 2
s = np.random.gamma(shape, scale, 1000)  # 1000개의 표본 생성

# 히스토그램
plt.hist(s, bins=30, density=True)
plt.title('Gamma Distribution')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()
```

```python
'''
베타 분포
베타 분포는 0과 1 사이의 값을 가지는 확률 변수를 모델링하는 데 사용됩니다. 예를 들어, 성공 확률을 모델링할 때 사용됩니다.

성공 확률: A/B 테스트에서 특정 버전이 성공할 확률.
투표율: 선거에서 특정 후보가 받을 투표율.
신뢰도: 제품의 신뢰도나 시스템의 가동률.
'''

# 베타 분포 예제
a, b = 2, 5  # a: 알파 매개변수 값 클수록 오른쪽, b: 베타 매개변수 값 작을수록 왼쪽
s = np.random.beta(a, b, 1000)  # 1000개의 표본 생성

# 히스토그램
plt.hist(s, bins=30, density=True)
plt.title('Beta Distribution')
plt.xlabel('Probability')
plt.ylabel('Frequency')
plt.show()
```

```python
'''
A/B 검정
왜 사용하는가? A/B 검정은 두 그룹 간의 차이를 비교하여 어떤 변화가 더 효과적인지 판단하는 데 사용됩니다.

실생활 예시:

웹사이트의 두 가지 디자인(A와 B) 중 어느 것이 더 많은 클릭을 유도하는지 비교.

가설검정 단계:

가설 설정: 귀무가설(H0)과 대립가설(H1)을 설정합니다.
H0: 두 그룹의 평균 클릭 수에 차이가 없다.
H1: 두 그룹의 평균 클릭 수에 차이가 있다.
유의수준 설정: 일반적으로 0.05를 사용합니다.
검정통계량 계산: t-검정을 사용하여 두 그룹의 평균을 비교합니다.
p-값 계산: 검정통계량에 대응하는 p-값을 계산합니다.
결론 도출: p-값이 유의수준보다 작으면 귀무가설을 기각합니다.
'''

import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# 예시 데이터 생성
np.random.seed(0)
data = pd.DataFrame({
    'Group': np.random.choice(['A', 'B'], size=100),
    'Clicks': np.random.randint(10, 50, size=100),
    'Views': np.random.randint(100, 200, size=100),
    'Conversions': np.random.randint(1, 10, size=100),
    'Time_Spent': np.random.uniform(1, 5, size=100),
    'Bounce_Rate': np.random.uniform(0.1, 0.5, size=100)
})

# 그룹별 클릭 수 비교
group_A = data[data['Group'] == 'A']['Clicks']
group_B = data[data['Group'] == 'B']['Clicks']

# t-검정을 사용하여 두 그룹의 평균 비교
t_stat, p_value = stats.ttest_ind(group_A, group_B)

print(f"t-statistic: {t_stat}, p-value: {p_value}")

# 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Clicks', data=data)
plt.title('Clicks by Group')
plt.show()
```

```python
'''
가설검정
왜 사용하는가? 가설검정은 특정 가설이 통계적으로 유의미한지 검증하는 방법입니다.

실생활 예시:

새로운 약이 기존 약보다 효과가 있는지 검증.

가설검정 단계:

가설 설정: 귀무가설(H0)과 대립가설(H1)을 설정합니다.
H0: 새로운 약과 기존 약의 평균 효과에 차이가 없다.
H1: 새로운 약의 평균 효과가 기존 약보다 크다.
유의수준 설정: 일반적으로 0.05를 사용합니다.
검정통계량 계산: t-검정을 사용하여 두 그룹의 평균을 비교합니다.
p-값 계산: 검정통계량에 대응하는 p-값을 계산합니다.
결론 도출: p-값이 유의수준보다 작으면 귀무가설을 기각합니다.
'''

# 예시 데이터 생성
np.random.seed(1)
data = pd.DataFrame({
    'Group': np.random.choice(['Existing', 'New'], size=100),
    'Effectiveness': np.random.normal(loc=5, scale=1, size=100)
})

# 그룹별 효과 비교
existing_drug = data[data['Group'] == 'Existing']['Effectiveness']
new_drug = data[data['Group'] == 'New']['Effectiveness']

# t-검정을 사용하여 두 그룹의 평균 비교
t_stat, p_value = stats.ttest_ind(existing_drug, new_drug)

print(f"t-statistic: {t_stat}, p-value: {p_value}")

# 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Effectiveness', data=data)
plt.title('Effectiveness by Drug Type')
plt.show()
```

```python
'''
재표본추출(bootstrap)은 작은 표본에서 신뢰구간을 추정하거나, 데이터의 분포를 이해하는 데 유용한 방법입니다.

실생활 예시:

고객 만족도 조사

한 회사가 고객 만족도 조사를 진행했습니다.
하지만, 전체 고객 중 일부만 설문에 응답했습니다.
이 작은 표본을 통해 전체 고객의 만족도를 추정하고 싶습니다.

재표본추출을 사용하는 이유: 작은 표본에서 얻은 데이터를 여러 번 재표본추출하여 신뢰구간을 추정하면,
전체 고객의 만족도를 더 정확하게 예측할 수 있습니다.

가설검정 단계:

가설 설정: 귀무가설(H0)과 대립가설(H1)을 설정합니다.
H0: 원본 데이터와 재표본추출 데이터의 평균에 차이가 없다.
H1: 원본 데이터와 재표본추출 데이터의 평균에 차이가 있다.
유의수준 설정: 일반적으로 0.05를 사용합니다.
검정통계량 계산: t-검정을 사용하여 두 그룹의 평균을 비교합니다.
p-값 계산: 검정통계량에 대응하는 p-값을 계산합니다.
결론 도출: p-값이 유의수준보다 작으면 귀무가설을 기각합니다.
'''

from sklearn.utils import resample

# 예시 데이터 생성
data = np.random.normal(loc=50, scale=10, size=100)

# 재표본추출
bootstrap_samples = resample(data, n_samples=100, replace=True)

# t-검정을 사용하여 원본 데이터와 재표본추출 데이터의 평균 비교
t_stat, p_value = stats.ttest_ind(data, bootstrap_samples)

print(f"Original Data Mean: {np.mean(data)}")
print(f"Bootstrap Sample Mean: {np.mean(bootstrap_samples)}")
print(f"t-statistic: {t_stat}, p-value: {p_value}")

# 시각화
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color='blue', label='Original Data')
sns.histplot(bootstrap_samples, kde=True, color='red', label='Bootstrap Sample')
plt.legend()
plt.title('Original Data vs Bootstrap Sample')
plt.show()
```

```python
'''
통계적 유의성과 p 값
왜 사용하는가? 통계적 유의성과 p 값은 결과의 유의성을 판단하는 기준입니다.

실생활 예시:

실험 결과가 우연에 의한 것인지 아닌지 판단.

가설검정 단계:

가설 설정: 귀무가설(H0)과 대립가설(H1)을 설정합니다.
H0: 두 그룹의 평균에 차이가 없다.
H1: 두 그룹의 평균에 차이가 있다.
유의수준 설정: 일반적으로 0.05를 사용합니다.
검정통계량 계산: t-검정을 사용하여 두 그룹의 평균을 비교합니다.
p-값 계산: 검정통계량에 대응하는 p-값을 계산합니다.
결론 도출: p-값이 유의수준보다 작으면 귀무가설을 기각합니다.
'''

# 예시 데이터 생성
np.random.seed(2)
data = pd.DataFrame({
    'Group': np.random.choice(['Control', 'Treatment'], size=100),
    'Outcome': np.random.normal(loc=5, scale=1, size=100)
})

# 그룹별 결과 비교
control_group = data[data['Group'] == 'Control']['Outcome']
treatment_group = data[data['Group'] == 'Treatment']['Outcome']

# t-검정을 사용하여 두 그룹의 평균 비교
t_stat, p_value = stats.ttest_ind(control_group, treatment_group)

print(f"t-statistic: {t_stat}, p-value: {p_value}")

# 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Outcome', data=data)
plt.title('Outcome by Group')
plt.show()
```

```python
'''
자유도란 무엇인가?
자유도는 데이터를 분석할 때, 독립적으로 변할 수 있는 값의 수를 의미합니다.
쉽게 말해, 데이터를 설명하는 데 필요한 정보의 양을 나타내는 것입니다.

실생활 예시: 친구들과의 피자 파티

친구들과 피자 파티를 열었습니다. 피자 한 판을 8조각으로 나눴고, 친구 7명과 함께 먹기로 했습니다.
각 친구가 몇 조각을 먹었는지 기록하려고 합니다.

자유도의 개념:

피자 8조각 중 7조각은 친구들이 몇 조각을 먹었는지 자유롭게 정할 수 있습니다.
마지막 1조각은 이미 7조각이 정해졌기 때문에 자동으로 결정됩니다.
즉, 8조각 중 7조각은 자유롭게 정할 수 있지만, 마지막 1조각은 이미 정해진 값에 따라 결정되기 때문에 자유도는 7입니다.

실생활 예시: 시험 점수 비교

두 반의 학생들이 시험을 봤습니다. 각 반의 학생 수는 10명이고, 두 반의 평균 점수를 비교하려고 합니다.

자유도의 개념:

각 반의 평균 점수를 계산할 때, 각 반의 점수 중 9개의 점수는 자유롭게 변할 수 있습니다.
마지막 1개의 점수는 이미 9개의 점수가 정해졌기 때문에 자동으로 결정됩니다.
따라서, 각 반의 자유도는 10 - 1 = 9이고, 두 반을 비교할 때의 총 자유도는 9 + 9 = 18입니다.

비유: 퍼즐 맞추기

퍼즐 조각이 10개 있는 퍼즐을 맞추려고 합니다.
9개의 퍼즐 조각은 자유롭게 맞출 수 있지만, 마지막 1개의 퍼즐 조각은 이미 9개의 조각이 맞춰졌기 때문에 자동으로 결정됩니다.


통계적 검정에서의 자유도

통계적 검정에서 자유도는 데이터를 분석할 때, 얼마나 많은 값이 자유롭게 변할 수 있는지를 나타냅니다.
예를 들어, t-검정에서 자유도는 표본 크기와 관련이 있습니다.

왜 자유도가 중요한가?
자유도는 통계적 검정에서 매우 중요한 역할을 합니다. 자유도가 높을수록 검정의 정확성이 높아지고, 결과를 더 신뢰할 수 있습니다. 자유도가 낮으면, 검정 결과가 불안정해질 수 있습니다.

이처럼, 자유도는 데이터를 분석할 때 자유롭게 변할 수 있는 값의 수를 나타내며, 이는 분석의 정확성과 신뢰성을 높이는 데 중요한 역할을 합니다.

자유도
왜 사용하는가? 자유도는 통계적 검정에서 사용되는 개념으로, 표본의 독립적인 정보의 수를 나타냅니다.

실생활 예시:

t 검정에서 자유도는 표본 크기와 관련.
'''

# 예시 데이터 생성
n = 10  # 표본 크기
degrees_of_freedom = n - 1

print(f"Degrees of Freedom: {degrees_of_freedom}")
```

```python
'''
t 검정
왜 사용하는가? t 검정은 두 그룹의 평균을 비교하는 데 사용됩니다.

실생활 예시:

두 가지 교육 방법의 효과를 비교.

가설검정 단계:

가설 설정: 귀무가설(H0)과 대립가설(H1)을 설정합니다.
H0: 두 교육 방법의 평균 점수에 차이가 없다.
H1: 두 교육 방법의 평균 점수에 차이가 있다.
유의수준 설정: 일반적으로 0.05를 사용합니다.
검정통계량 계산: t-검정을 사용하여 두 그룹의 평균을 비교합니다.
p-값 계산: 검정통계량에 대응하는 p-값을 계산합니다.
결론 도출: p-값이 유의수준보다 작으면 귀무가설을 기각합니다
'''

# 예시 데이터 생성
np.random.seed(3)
data = pd.DataFrame({
    'Group': np.random.choice(['Method A', 'Method B'], size=100),
    'Scores': np.random.normal(loc=75, scale=10, size=100)
})

# 그룹별 점수 비교
method_A = data[data['Group'] == 'Method A']['Scores']
method_B = data[data['Group'] == 'Method B']['Scores']

# t-검정을 사용하여 두 그룹의 평균 비교
t_stat, p_value = stats.ttest_ind(method_A, method_B)

print(f"t-statistic: {t_stat}, p-value: {p_value}")

# 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Scores', data=data)
plt.title('Scores by Method')
plt.show()
```

```python
'''
다중검정은 여러 가설을 동시에 검정할 때 발생하는 문제를 해결하기 위해 사용됩니다.
여러 번의 검정을 수행하면 유의수준이 증가하여 잘못된 결론을 내릴 확률이 높아지기 때문에, 이를 보정하는 방법이 필요합니다.

유의수준 증가의 원리

단일 검정:

한 번의 검정을 수행할 때, 유의수준이 0.05라면, 5%의 확률로 잘못된 결론을 내릴 수 있습니다.
다중검정:

만약 20번의 독립적인 검정을 수행한다면, 각 검정에서 5%의 확률로 잘못된 결론을 내릴 수 있습니다.
이 경우, 적어도 한 번 이상 잘못된 결론을 내릴 확률은 (1 - (1 - 0.05)^{20} \approx 0.64)로 약 64%가 됩니다.
즉, 검정의 수가 많아질수록 잘못된 결론을 내릴 확률이 급격히 증가하게 됩니다. 이를 다중검정 문제라고 합니다.

실생활 예시: 여러 약물의 효과 검증
상황: 한 제약회사가 다섯 가지 새로운 약물(A, B, C, D, E)의 효과를 검증하고자 합니다.
각 약물의 효과를 기존 약물과 비교하여 유의미한 차이가 있는지 확인하려고 합니다.

가설검정 단계:

가설 설정:
귀무가설(H0): 각 약물의 평균 효과에 차이가 없다.
대립가설(H1): 각 약물의 평균 효과에 차이가 있다.
유의수준 설정: 일반적으로 0.05를 사용합니다.
검정통계량 계산: 각 약물에 대해 t-검정을 수행합니다.
p-값 계산: 각 검정통계량에 대응하는 p-값을 계산합니다.
다중검정 보정: 다중검정 보정을 통해 유의수준을 조정합니다.
결론 도출: 보정된 p-값이 유의수준보다 작으면 귀무가설을 기각합니다.

다중검정 보정의 원리
다중검정 보정은 여러 가설을 동시에 검정할 때 발생하는 유의수준 증가 문제를 해결하기 위한 방법입니다.
대표적인 보정 방법으로는 본페로니 보정과 홀름-본페로니 보정이 있습니다.

본페로니 보정 (Bonferroni Correction):
각 검정의 유의수준을 전체 유의수준을 검정의 수로 나누어 조정합니다.
예를 들어, 5개의 검정을 수행하고 전체 유의수준을 0.05로 설정한 경우, 각 검정의 유의수준은 0.05/5 = 0.01이 됩니다.

홀름-본페로니 보정 (Holm-Bonferroni Correction):
본페로니 보정의 단점을 보완한 방법으로, p-값을 오름차순으로 정렬한 후 순차적으로 보정합니다.
첫 번째 p-값은 전체 유의수준을 검정의 수로 나눈 값과 비교하고, 두 번째 p-값은 전체 유의수준을 (검정의 수 - 1)로 나눈 값과 비교하는 방식입니다.
'''

from statsmodels.stats.multitest import multipletests

# 예시 데이터 생성
np.random.seed(4)
data = pd.DataFrame({
    'Drug': np.random.choice(['Drug A', 'Drug B', 'Drug C', 'Drug D', 'Drug E'], size=100),
    'Effectiveness': np.random.normal(loc=5, scale=1, size=100)
})

# 각 약물의 효과 비교
p_values = []
for drug in data['Drug'].unique():
    drug_data = data[data['Drug'] == drug]['Effectiveness']
    t_stat, p_value = stats.ttest_1samp(drug_data, 5)
    p_values.append(p_value)

# 다중검정 수행
reject, pvals_corrected, _, _ = multipletests(p_values, alpha=0.05, method='bonferroni')

print(f"Significant Results: {reject}")
print(f"Corrected p-values: {pvals_corrected}")

# 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(x='Drug', y='Effectiveness', data=data)
plt.title('Effectiveness by Drug')
plt.show()
```

```python
'''
분산분석
왜 사용하는가? 분산분석은 여러 그룹 간의 차이를 비교하는 데 사용됩니다.

실생활 예시:

여러 교육 방법의 효과를 비교.

가설검정 단계:

가설 설정: 귀무가설(H0)과 대립가설(H1)을 설정합니다.
H0: 여러 그룹의 평균에 차이가 없다.
H1: 적어도 하나의 그룹의 평균이 다른 그룹과 차이가 있다.
유의수준 설정: 일반적으로 0.05를 사용합니다.
검정통계량 계산: F-검정을 사용하여 여러 그룹의 평균을 비교합니다.
p-값 계산: 검정통계량에 대응하는 p-값을 계산합니다.
결론 도출: p-값이 유의수준보다 작으면 귀무가설을 기각합니다.
'''

from scipy.stats import f_oneway

# 예시 데이터 생성
np.random.seed(5)
data = pd.DataFrame({
    'Group': np.random.choice(['Group 1', 'Group 2', 'Group 3'], size=100),
    'Scores': np.random.normal(loc=75, scale=10, size=100)
})

# 세 그룹의 점수 데이터
group1 = data[data['Group'] == 'Group 1']['Scores']
group2 = data[data['Group'] == 'Group 2']['Scores']
group3 = data[data['Group'] == 'Group 3']['Scores']

# 분산분석 수행
f_stat, p_value = f_oneway(group1, group2, group3)

print(f"F-statistic: {f_stat}, p-value: {p_value}")

# 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Scores', data=data)
plt.title('Scores by Group')
plt.show()
```

```python
'''
카이제곱 검정의 원리
카이제곱 검정은 관찰된 빈도와 기대된 빈도 간의 차이를 비교하여 두 범주형 변수 간의 독립성을 검증합니다. 이를 통해 두 변수 간의 관계가 있는지 여부를 판단할 수 있습니다.

가설검정 단계:
가설 설정:
귀무가설(H0): 두 범주형 변수 간에 독립성이 있다.
대립가설(H1): 두 범주형 변수 간에 독립성이 없다.
유의수준 설정: 일반적으로 0.05를 사용합니다.
검정통계량 계산: 카이제곱 통계량을 계산합니다.
카이제곱 통계량은 다음과 같이 계산됩니다:
관찰된 빈도(O)와 기대된 빈도(E)의 차이를 제곱한 후, 이를 기대된 빈도로 나누어 계산합니다.
모든 범주에 대해 이 값을 합산하여 최종 카이제곱 통계량을 구합니다.
p-값 계산: 검정통계량에 대응하는 p-값을 계산합니다.
결론 도출: p-값이 유의수준보다 작으면 귀무가설을 기각합니다.
실생활 예시: 고객 만족도와 재구매 의사
상황: 한 회사가 고객 만족도와 재구매 의사 간의 관계를 조사하고자 합니다. 고객 만족도는 “만족”, “보통”, "불만족"으로 분류되고, 재구매 의사는 “예”, "아니오"로 분류됩니다.

가설검정 단계:

가설 설정:
귀무가설(H0): 고객 만족도와 재구매 의사 간에 독립성이 있다.
대립가설(H1): 고객 만족도와 재구매 의사 간에 독립성이 없다.
유의수준 설정: 0.05
검정통계량 계산: 카이제곱 통계량을 계산합니다.
p-값 계산: 검정통계량에 대응하는 p-값을 계산합니다.
결론 도출: p-값이 유의수준보다 작으면 귀무가설을 기각합니다.


단계별 원리 적용
가설 설정:
귀무가설(H0): 고객 만족도와 재구매 의사 간에 독립성이 있다.
대립가설(H1): 고객 만족도와 재구매 의사 간에 독립성이 없다.

유의수준 설정:
유의수준을 0.05로 설정합니다. 이는 5%의 확률로 잘못된 결론을 내릴 수 있다는 의미입니다.

검정통계량 계산:
관찰된 빈도와 기대된 빈도를 비교하여 카이제곱 통계량을 계산합니다.
예를 들어, 다음과 같은 관찰된 빈도 데이터가 있다고 가정합니다:

        만족   보통   불만족
예       30     20     10
아니오   10     20     30

기대된 빈도는 각 행과 열의 합을 이용하여 계산합니다. 예를 들어, "만족"과 "예"의 기대된 빈도는 다음과 같이 계산됩니다:
기대된 빈도 = (행의 합) * (열의 합) / 전체 합 = 60 * 40 / 120 = 20

p-값 계산:
카이제곱 통계량을 이용하여 p-값을 계산합니다. p-값은 관찰된 데이터가 귀무가설 하에서 나타날 확률을 의미합니다.

결론 도출:
p-값이 유의수준(0.05)보다 작으면 귀무가설을 기각합니다. 이는 고객 만족도와 재구매 의사 간에 독립성이 없다는 것을 의미합니다.


'''

from scipy.stats import chi2_contingency

# 예시 데이터 생성
np.random.seed(6)
data = pd.DataFrame({
    'Gender': np.random.choice(['Male', 'Female'], size=100),
    'Preference': np.random.choice(['Product A', 'Product B'], size=100)
})

# 관찰된 빈도 데이터
contingency_table = pd.crosstab(data['Gender'], data['Preference'])

# 카이제곱검정 수행
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-squared statistic: {chi2}, p-value: {p}, Degrees of Freedom: {dof}")
print(f"Expected Frequencies: \n{expected}")

# 시각화
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, cmap='coolwarm')
plt.title('Contingency Table: Gender vs Preference')
plt.show()
```

```python
'''
멀티암드 밴딧 알고리즘(Multi-Armed Bandit Algorithm)은 여러 선택지 중 최적의 선택을 찾는 문제를 해결하는 데 사용됩니다.
이 알고리즘은 탐색(exploration)과 활용(exploitation) 사이의 균형을 맞추는 것이 핵심입니다. 이제 이 알고리즘의 원리를 실생활 예시를 통해 상세하게 설명해드릴게요.

멀티암드 밴딧 알고리즘의 원리
멀티암드 밴딧 문제는 여러 개의 슬롯 머신(밴딧) 중에서 어느 슬롯 머신을 당겨야 가장 많은 보상을 받을 수 있는지를 결정하는 문제입니다.
각 슬롯 머신은 서로 다른 보상 확률을 가지고 있으며, 이를 알지 못한 상태에서 최적의 선택을 찾아야 합니다.

탐색과 활용의 균형
탐색(Exploration): 새로운 선택지를 시도하여 보상 확률을 알아내는 과정입니다.
활용(Exploitation): 현재까지 얻은 정보를 바탕으로 가장 높은 보상을 줄 것으로 예상되는 선택지를 선택하는 과정입니다.
실생활 예시: 광고 배너의 최적화
상황: 한 웹사이트에서 여러 개의 광고 배너를 테스트하고 있습니다. 각 광고 배너는 클릭될 확률이 다릅니다. 목표는 가장 높은 클릭률을 가진 광고 배너를 찾아내는 것입니다.

단계별 원리 적용
가설 설정:
각 광고 배너의 클릭 확률을 추정합니다.
유의수준 설정:
일반적으로 0.05를 사용합니다. 이는 5%의 확률로 잘못된 결론을 내릴 수 있다는 의미입니다.
검정통계량 계산:
각 광고 배너의 클릭 횟수와 노출 횟수를 기록합니다.
예를 들어, 광고 배너 A가 100번 노출되어 10번 클릭되었다면, 클릭 확률은 10/100 = 0.1입니다.
p-값 계산:
각 광고 배너의 클릭 확률을 비교하여 p-값을 계산합니다. p-값은 두 광고 배너의 클릭 확률이 유의미하게 다른지 여부를 판단하는 데 사용됩니다.
결론 도출:
p-값이 유의수준(0.05)보다 작으면, 두 광고 배너의 클릭 확률이 유의미하게 다르다고 결론짓습니다.
알고리즘의 구체적인 방법
멀티암드 밴딧 알고리즘에는 여러 가지 변형이 있습니다. 그 중 대표적인 방법 두 가지를 소개합니다.

엡실론-그리디 알고리즘 (Epsilon-Greedy Algorithm):
탐색: 엡실론(ε)의 확률로 무작위로 선택합니다.
활용: 1-ε의 확률로 현재까지 가장 높은 보상을 준 선택지를 선택합니다.
예를 들어, ε = 0.1로 설정하면, 10%의 확률로 무작위 선택을 하고, 90%의 확률로 현재까지 가장 높은 클릭률을 보인 광고 배너를 선택합니다.
'''
import numpy as np

# 광고 배너의 클릭 확률 (예시) 각 광고 배너의 실제 클릭 확률을 설정합니다.
true_click_rates = [0.05, 0.10, 0.15]

# 엡실론-그리디 알고리즘 함수 정의
def epsilon_greedy(epsilon, n_trials):
    n_arms = len(true_click_rates)  # 광고 배너의 개수를 설정합니다.
    counts = np.zeros(n_arms)  # 각 광고 배너의 클릭 횟수를 저장할 배열을 0으로 초기화합니다.
    rewards = np.zeros(n_arms)  # 각 광고 배너의 보상(클릭) 횟수를 저장할 배열을 0으로 초기화합니다.

    for _ in range(n_trials):  # n_trials 횟수만큼 반복합니다.
        if np.random.rand() < epsilon:  # 무작위 값이엡실론보다 작으면 탐색을 수행합니다.
            arm = np.random.randint(n_arms)  # 무작위로 광고 배너를 선택합니다. (탐색)
        else:
            arm = np.argmax(rewards / (counts + 1e-5))  # 현재까지의 보상 횟수를 기반으로 가장 높은 보상을 준 광고 배너를 선택합니다. (활용)

        reward = np.random.rand() < true_click_rates[arm]  # 선택한 광고 배너의 클릭 여부를 결정합니다.
        counts[arm] += 1  # 선택한 광고 배너의 클릭 횟수를 증가시킵니다.
        rewards[arm] += reward  # 선택한 광고 배너의 보상(클릭) 횟수를 증가시킵니다.

    return counts, rewards  # 각 광고 배너의 클릭 횟수와 보상 횟수를 반환합니다.

epsilon = 0.1  #엡실론 값을 설정합니다.
n_trials = 1000  # 시도 횟수를 설정합니다.
counts, rewards = epsilon_greedy(epsilon, n_trials)  #엡실론-그리디 알고리즘을 실행하여 결과를 얻습니다.

print(f"Counts: {counts}")  # 각 광고 배너의 클릭 횟수를 출력합니다.
print(f"Rewards: {rewards}")  # 각 광고 배너의 보상(클릭) 횟수를 출력합니다.
print(f"Estimated Click Rates: {rewards / counts}")  # 각 광고 배너의 추정 클릭 확률을 출력합니다.
```

```python
'''
검정력과 표본크기
왜 사용하는가? 검정력은 검정의 민감도를 나타내며, 표본크기는 필요한 표본의 크기를 결정하는 데 사용됩니다.

실생활 예시:

임상 시험에서 필요한 표본 크기 결정.

가설검정 단계:

가설 설정: 귀무가설(H0)과 대립가설(H1)을 설정합니다.
H0: 두 그룹의 평균에 차이가 없다.
H1: 두 그룹의 평균에 차이가 있다.
유의수준 설정: 일반적으로 0.05를 사용합니다.
검정통계량 계산: t-검정을 사용하여 두 그룹의 평균을 비교합니다.
p-값 계산: 검정통계량에 대응하는 p-값을 계산합니다.
결론 도출: p-값이 유의수준보다 작으면 귀무가설을 기각합니다.
검정력 계산: 검정력 분석을 통해 필요한 표본 크기를 결정합니다.



작은 효과 크기 (d = 0.2)
상황: 새로운 의약품 A가 기존 의약품 B보다 조금 더 효과적이라는 가설을 검토하고 있습니다. 효과 크기가 0.2로 측정되었습니다.
표본 크기 결정: 작은 효과를 신뢰할 수 있는 결과로 판단하기 위해 대규모의 표본이 필요합니다. 예를 들어, 500명의 환자가 필요할 수 있습니다.


중간 효과 크기 (d = 0.5)
상황: 의약품 A가 B보다 확실히 더 효과적이라고 가정합니다. 효과 크기가 0.5로 측정되었습니다.
표본 크기 결정: 중간 정도의 효과를 검토하기 위해 적절한 표본 크기로 200명을 사용할 수 있습니다.


큰 효과 크기 (d = 0.8)
상황: 의약품 A가 B보다 훨씬 더 효과적이라고 가정합니다. 효과 크기가 0.8로 측정되었습니다.
표본 크기 결정: 큰 효과를 검출하기 위해 적은 수의 표본으로도 충분하므로, 50명의 환자만으로도 충분할 수 있습니다.



효과 크기 (Effect Size): 두 의약품의 효과 차이를 나타내며, 효과 크기가 클수록 차이가 크고 뚜렷합니다.
표본 크기 (Sample Size): 효과 크기에 따라 필요한 데이터 양이 달라집니다. 효과 크기가 클수록 표본 크기는 작아져도 됩니다.

'''

from statsmodels.stats.power import TTestIndPower

# 효과 크기, 유의수준, 검정력
effect_size = 0.5
alpha = 0.05
power = 0.8

# 필요한 표본 크기 계산
analysis = TTestIndPower()
sample_size = analysis.solve_power(effect_size, power=power, alpha=alpha)

print(f"Required Sample Size: {sample_size}")

# 시각화
effect_sizes = np.linspace(0.1, 1.0, 50)
sample_sizes = [analysis.solve_power(effect_size=es, power=power, alpha=alpha) for es in effect_sizes]

plt.figure(figsize=(10, 6))
plt.plot(effect_sizes, sample_sizes, marker='o')
plt.xlabel('Effect Size')
plt.ylabel('Sample Size')
plt.title('Sample Size vs Effect Size')
plt.grid(True)
plt.show()
```
